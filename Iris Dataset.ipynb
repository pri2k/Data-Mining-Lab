```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
# Load iris dataset from sklearn
iris = datasets.load_iris()
# print(iris)
print(iris.keys())
print(iris.feature_names)
print(iris.target_names)

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
iris_df['species'] = iris_df['species'].map(dict(zip(range(3), iris.target_names)))

print()
print("First 5 rows of Iris dataset:")
print(iris_df.head())
```

```python
# Summary statistics
print("\nSummary Statistics for Iris:")
print(iris_df.describe())

# Check for missing values
print("\nMissing values in Iris:")
print(iris_df.isnull().sum())

# Class distribution
print("\nSpecies distribution:")
print(iris_df['species'].value_counts())
```

```python
# Histograms
iris_df.hist(figsize=(10, 8), edgecolor='black')
plt.suptitle('Histogram of Iris Features')
plt.show()

# Scatter plot
sns.pairplot(iris_df, hue='species')
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

```python
print("\nSummary Report:")

# Iris
print(f"- Iris dataset has {iris_df.shape[0]} rows and {iris_df.shape[1]} columns.")
print(f"- Features: {list(iris_df.columns[:-1])}")
print(f"- Classes: {iris.target_names.tolist()}")
```
